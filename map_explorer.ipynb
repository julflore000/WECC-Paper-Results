```python
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = Dataset('wecc_results.nc')
regions = ['basin','california','mountains','northwest','southwest']
years = np.array([2016, 2017, 2018]).astype(str)
capacities = ['100 MW','1 GW','5 GW']
technologies = ['solar','wind']
add_ons = ['',' 500 MW Storage','2x Renewables']
```

```python
# map options
cmap = 'plasma'
cmap_diff = 'coolwarm'
```

Find average difference due to generator size...

Solar:

```python
average_diff = []
technology = 'solar'
for region in regions:
    for year in years:

        solar_100 = d.variables[' '.join([region,year,'100 MW', technology])][:]
        solar_1000 = d.variables[' '.join([region,year,'1 GW', technology])][:]
        average_diff.append(np.average(solar_1000-solar_100))

        if region == 'basin' and int(year) == 2016:
            fig, ax = plt.subplots(1,3,figsize=(30,10))
            ax[0].imshow(solar_100,cmap=cmap,vmin=0,vmax=100)
            ax[1].imshow(solar_1000,cmap=cmap,vmin=0,vmax=100)
            ax[2].imshow(solar_100-solar_1000,cmap=cmap_diff,vmin=-15,vmax=15)

            ax[0].set_title('100 MW Solar')
            ax[1].set_title('1000 MW Solar')
            ax[2].set_title('$\Delta 100 -> 1000 Solar$')


print('Average Difference in Solar Capacity Values from 100 MW Generators and 1 GW generators:')
print(np.average(average_diff))

    
   
    
```
